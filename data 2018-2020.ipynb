```python
import os
import pandas as pd
import calendar

# Directory containing CSV files
directory_2018 = 'data/2018/'

# List to store dataframes
dfs_2018 = []

# Iterate over CSV files in the directory
for idx, filename in enumerate(os.listdir(directory_2018)):
    if filename.endswith('.csv'):
        # Read CSV file
        df = pd.read_csv(os.path.join(directory_2018, filename))
        
        # Add month column with the name of the month
        month_name = calendar.month_name[idx + 1]
        df['month'] = month_name
        
        # Append dataframe to list
        dfs_2018.append(df)

# Concatenate or merge all dataframes in the list
merged_df_2018 = pd.concat(dfs_2018, ignore_index=True)

# Display merged dataframe
merged_df_2018.head()
```

```python
merged_df_2018["year"]=2018
merged_df_2018.head()
```
